# Plotting x and y points
The plot() function is used to draw points (markers) in a diagram.

By default, the plot() function draws a line from point to point.

```python
import matplotlib.pyplot as plt
import numpy as np

x=[1,5,8]
y=[1,7,10]

#plt.plot(x, y)
plt.plot(x, y,ls = '--' )

plt.show()
```

# Plotting Without Line
To plot only the markers, you can use shortcut string notation parameter 'o', which means 'rings'.

```python
import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([2, 8])
ypoints = np.array([2, 8])

plt.plot(xpoints, ypoints, 'o') 
plt.show()
```

```python
#You can plot as many points as you like, just make sure you have the same number of points in both axis.
```

```python
import matplotlib.pyplot as plt
import numpy as np

x=[2,5,8]
y=[2,6,9]
plt.plot(x, y, '*' )
#plt.plot(x, y, '*' , ms = 20 , c='w', mec = 'r' , mfc = 'y') # scatter plot with marker *

plt.title("Scatter Plot", loc = 'center') #left, right, center
plt.xlabel("X-Axis")
plt.ylabel("Y-axis")
plt.grid()
plt.show()
```

# Marker Properties

1. argument : markersize or the shorter version, ms

   plt.plot(ypoints, marker = 'o', ms = 20)
   
   
2. argument : markeredgecolor or the shorter mec

   plt.plot(ypoints, marker = 'o', ms = 20, mec = 'r')
   
   
3. argument : markerfacecolor or the shorter mfc to set the color inside the edge of the markers:

   plt.plot(ypoints, marker = 'o', ms = 20, mfc = 'r')
   
NB: Use both the mec and mfc arguments to color of the entire marker:

plt.plot(ypoints, marker = 'o', ms = 20, mec = 'r', mfc = 'r') 

plt.plot(ypoints, marker = 'o', ms = 20, mec = '#4CAF50', mfc = '#4CAF50')

plt.plot(ypoints, marker = 'o', ms = 20, mec = 'hotpink', mfc = 'hotpink')


4. Line style : linestyle = 'dotted' or ls = ':' and linewidth = '20.5'

    plt.plot(ypoints, linestyle = 'dotted')
    
    'solid' (default)	'-'	
    'dotted'	':'	
    'dashed'	'--'	
    'dashdot'	'-.'	
    'None'	'' or ' '

# Label, Title, Font and Grid

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

font1 = {'family':'serif','color':'blue','size':18}
font2 = {'family':'serif','color':'darkred','size':15}

plt.title("Sports Watch Data", fontdict = font1, loc = 'center') #left, right, center
plt.xlabel("Average Pulse", fontdict = font2)
plt.ylabel("Calorie Burnage", fontdict = font2)

plt.plot(x, y)
plt.grid() # axis = 'y', axis='x'
plt.show()
```

# Scatter Plots

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([1,2,3,4,5])
y = np.array([5,10,15,15,20])
plt.scatter(x, y, color='hotpink')

plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([1,2,3,4,5])
y = np.array([5,10,15,15,20])
plt.scatter(x, y, c='hotpink')

x = np.array([2,4,6,8,10])
y = np.array([5,7,10,12,18])
plt.scatter(x, y, color = '#88c999')

plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([1,2,3,4,5])
y = np.array([10,20,30,40,50])

colors = np.array([0, 10, 20, 30, 40])
plt.scatter(x, y, c=colors, cmap='viridis')

plt.colorbar()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([5,7,8,9,2])
y = np.array([99,86,87,88,11])
sizes = np.array([20,50,100,200,500])

plt.scatter(x, y, s=sizes, alpha=0.5)

plt.show()
```

# Pie Charts

```python
import matplotlib.pyplot as plt
import numpy as np

x=[10,15,25]

plt.pie(x)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([35, 25, 25, 15])
l = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(x, labels = l)

plt.show() 
```

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels, startangle = 90)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

data = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.2, 0, 0, 0]

plt.pie(data, labels = mylabels, explode = myexplode)
plt.legend(title='Four Fruits')
plt.show() 
```

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
mycolors = ["black", "hotpink", "b", "#4CAF50"]

plt.pie(y, labels = mylabels, colors = mycolors)
plt.show() 
```
